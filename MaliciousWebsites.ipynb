```python
data = pd.read_csv("/content/drive/MyDrive/PDFMalware2022.csv")
data = data.dropna()
print(len(data))

print(data.columns)
```

```python
types = []
for i in data['Class']:
  if i=='Malicious':
    types.append(1)
  else:
    types.append(0)

data["Type"] = types

print(data["Type"])
corr  = data.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
for i in corr:
  if(corr["Type"][i]<-0.12):
    print(i)
```

```python
training = data.iloc[:7016, :]
test = data.iloc[7016:, :]
print(len(training))
print(len(test))

```

```python
logr = linear_model.LogisticRegression()
stream = np.array(training["pages"]).reshape(-1, 1)
types = np.array(training["Type"]).reshape(-1, 1)
logr.fit(stream,types)
```

```python
test_stream = np.array(test["pages"])#.reshape(-1, 1)
print(np.shape(test_stream))
test_stream = np.array(test["pages"]).reshape(-1, 1)
print(np.shape(test_stream))
y_actual = np.array(test["Type"]).reshape(-1, 1)
prediction = []
y_pred = logr.predict(test_stream)
print(y_pred)
print(y_actual)
test_stream = np.array(test["pages"]).reshape(3007,)
print(np.shape(test_stream))
```

```python
#evaluation metrics

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_actual, y_pred)
print(cm)
```
